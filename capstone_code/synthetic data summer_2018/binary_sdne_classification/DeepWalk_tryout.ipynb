```python
import nilearn as nl
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import pickle as pkl
import time
import os
import math
import csv
```

```python
data_t = pkl.load(open( "data.p", "rb" ))
trn_x = data_t['trn_x']
trn_y = data_t['trn_y']
val_x = data_t['val_x']
val_y = data_t['val_y']

num_trn = trn_x.shape[0]
num_val = val_x.shape[0]
num_node = 50

print(num_trn, num_val)

sdne_data = {}
sdne_data['trn_x'] = np.zeros((num_trn, 64))
sdne_data['trn_y'] = trn_y
sdne_data['val_x'] = np.zeros((num_val, 64))
sdne_data['val_y'] = val_y

for i in range(num_trn):
    print('trn ', i, flush=True)
    file = open('adjlist.txt', 'w')
    trn_x_sample = trn_x[i]
    trn_x_sample = trn_x_sample.reshape(num_node+1, num_node+1)
    for j in range(num_node+1):
        line = trn_x_sample[j, ]
        index = np.where(line > 0)
        index = index[0]
        res = np.append(j, index)
        res = ' '.join(str(x) for x in res)
        file.write(res)
        file.write('\n')
    file.close()
    os.system('python -m openne --method sdne --input adjlist.txt --graph-format adjlist --output out.txt --bs 1 --encoder-list [1000,64] --lr 0.001')
    temp = []
    with open('out.txt', newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=' ', quoting=csv.QUOTE_NONNUMERIC)
        for row in reader:
            temp.append(row)
    temp = temp[1:]
    temp = sorted(temp, key=lambda x: x[0])
    new_sample = []
    print(temp[-1], flush=True)
    new_sample = np.asarray(temp[-1][1:])
    new_sample = new_sample.reshape(1, 64)
    sdne_data['trn_x'][i] = new_sample


for i in range(num_val):
    print('val ', i, flush=True)
    file = open('adjlist.txt', 'w')
    val_x_sample = val_x[i]
    val_x_sample = val_x_sample.reshape(num_node+1, num_node+1)
    for j in range(num_node+1):
        line = val_x_sample[j, ]
        index = np.where(line > 0)
        index = index[0]
        res = np.append(j, index)
        res = ' '.join(str(x) for x in res)
        file.write(res)
        file.write('\n')
    file.close()
    os.system('python -m openne --method sdne --input adjlist.txt --graph-format adjlist --output out.txt --bs 1 --encoder-list [1000,64] --lr 0.001')
    temp = []
    with open('out.txt', newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=' ', quoting=csv.QUOTE_NONNUMERIC)
        for row in reader:
            temp.append(row)
    temp = temp[1:]
    temp = sorted(temp, key=lambda x: x[0])
    new_sample = []
    print(temp[-1], flush=True)
    new_sample = np.asarray(temp[-1][1:])
    new_sample = new_sample.reshape(1, 64)
    sdne_data['val_x'][i] = new_sample
    
pkl.dump(sdne_data, open(os.path.join('.', 'sdne_data.p'), 'wb'))
```
